```python
# %matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# File to Load (Remember to change these)
city_data_to_load = "data/city_data.csv"
ride_data_to_load = "data/ride_data.csv"

# Read the City and Ride Data
city_data = pd.read_csv(city_data_to_load)
ride_data = pd.read_csv(ride_data_to_load)

# Combine the data into a single dataset
df = pd.merge(ride_data, city_data, on="city", how='left')
# Display the data table for preview
df.head()
```

```python
columns = ['City', 'Type', 'Number of drivers', 'Average fare', 'Total rides']
city_df = pd.DataFrame(columns=columns)
city_df["City"] = df["city"]
city_df.set_index("City")
```

## Bubble Plot of Ride Sharing Data

```python
# Obtain the x and y coordinates for each of the three city types
x_axis = df.groupby("city")["type"].value_counts()
x_axis_urban = df.loc[df['type'] == 'Urban'].groupby("city")["type"].value_counts()
x_axis_suburban = df.loc[df['type'] == "Suburban"].groupby("city")["type"].value_counts()
x_axis_rural = df.loc[df['type'] == 'Rural'].groupby("city")["type"].value_counts()

y_axis = df.groupby("type")["fare"].mean()
y_axis_urban = df.loc[df['type'] == 'Urban'].groupby("city")["fare"].mean()
y_axis_suburban = df.loc[df['type'] == 'Suburban'].groupby("city")["fare"].mean()
y_axis_rural = df.loc[df['type'] == 'Rural'].groupby("city")["fare"].mean()

driver_per_city = df.groupby("city")['driver_count'].sum()
# Build the scatter plots for each city types
# sizes = driver_per_city
plt.scatter(x_axis_urban, y_axis_urban, marker="o", facecolors="coral", edgecolors="black",
            s=driver_per_city/10, alpha=0.75)
plt.scatter(x_axis_suburban, y_axis_suburban, marker="o", facecolors="skyblue", edgecolors="black",
            s=driver_per_city/10, alpha=0.75)
plt.scatter(x_axis_rural, y_axis_rural, marker="o", facecolors="yellow", edgecolors="black",
            s=driver_per_city/10, alpha=0.75)

# Incorporate the other graph properties
plt.xlim(x_axis.min() - 2,x_axis.max() + 2)
plt.ylim(y_axis.min() - 5,y_axis.max() + 10)
plt.grid()
# Create a legend

# Incorporate a text label regarding circle size

# Save Figure
```

## Total Fares by City Type

```python
# Calculate Type Percents
city_type = ['Rural', 'Suburban', 'Urban']
type_count = df.groupby("type")["fare"].sum()
colors = ['orange', 'lightblue' ,'lightcoral' ,]
explode = [0, 0, .10]
# Build Pie Chart
plt.pie(type_count, explode=explode, labels=city_type, colors=colors,
        autopct="%1.1f%%", shadow=True, startangle = 150)
plt.show()
# Save Figure
plt.savefig("../Images/TotalFaresByCityType.png")

```
