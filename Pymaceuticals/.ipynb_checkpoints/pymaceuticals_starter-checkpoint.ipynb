```python
# Dependencies and Setup
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import scipy
from scipy import stats

# Hide warning messages in notebook
import warnings
warnings.filterwarnings('ignore')

# File to Load (Remember to Change These)
mouse_drug_data_to_load = "data/mouse_drug_data.csv"
clinical_trial_data_to_load = "data/clinicaltrial_data.csv"

# Read the Mouse and Drug Data and the Clinical Trial Data
mouse_drug_data_df = pd.read_csv(mouse_drug_data_to_load)
clinical_trial_data_df = pd.read_csv(clinical_trial_data_to_load)

# Combine the data into a single dataset
df = pd.merge(clinical_trial_data_df, mouse_drug_data_df, on="Mouse ID", how="left" )

# Display the data table for preview
df.head()
```

## Tumor Response to Treatment

```python
# Store the Mean Tumor Volume Data Grouped by Drug and Timepoint 
mean_data = df.groupby(["Drug", "Timepoint"]).mean()["Tumor Volume (mm3)"]

# Convert to DataFrame
mean_df = pd.DataFrame({'Tumor Volume (mm3)' : mean_data}).reset_index()

# Preview DataFrame
mean_df.head()
```

```python
# Store the Standard Error of Tumor Volumes Grouped by Drug and Timepoint
tumor_vol_sem = df.groupby(["Drug", "Timepoint"]).sem()["Tumor Volume (mm3)"]

# Convert to DataFrame
tumor_vol_sem_df = pd.DataFrame(tumor_vol_sem).reset_index()
# Preview DataFrame
tumor_vol_sem_df.head()
```

```python
# Minor Data Munging to Re-Format the Data Frames
mean_df = mean_df.reset_index()
pivoted = mean_df.pivot(index='Timepoint', columns='Drug', values='Tumor Volume (mm3)')

# Preview that Reformatting worked
pivoted.head()
```

```python
# Generate the Plot (with Error Bars)

plt.errorbar(pivoted.index, 
             pivoted['Capomulin'], 
             yerr=tumor_vol_sem['Capomulin'].values, 
             linewidth=0.5, 
             marker="o",
             color='r',
             dashes=[2, 1]
            )
plt.errorbar(pivoted.index, 
             pivoted['Infubinol'], 
             yerr=tumor_vol_sem['Infubinol'].values, 
             linewidth=0.5, 
             marker="^",
             color='b',
             dashes=[2, 1]
            )
plt.errorbar(pivoted.index, 
             pivoted['Ketapril'], 
             yerr=tumor_vol_sem['Ketapril'].values, 
             linewidth=0.5, 
             marker="s",
             color='g',
             dashes=[2, 1]
            )
plt.errorbar(pivoted.index, 
             pivoted['Placebo'], 
             yerr=tumor_vol_sem['Placebo'].values, 
             linewidth=0.5, 
             marker="D",
             color='k',
             dashes=[2, 1]
            )

plt.grid()
plt.title('Tumor Response to Treatment')
plt.xlabel('Time (Days)')
plt.ylabel('Tumor Volume (mm3)')
plt.legend(loc="best", fontsize="small")
# Save the Figure
plt.savefig("TumorVolvsTime.png", bbox_inches = 'tight')
```

```python
# Show the Figure
plt.show()
```

## Metastatic Response to Treatment

```python
# Store the Mean Met. Site Data Grouped by Drug and Timepoint 
mean_met = df.groupby(["Drug", "Timepoint"]).mean()["Metastatic Sites"]

# Convert to DataFrame
met_df = pd.DataFrame({'Metastatic Sites' : mean_met}).reset_index()

# Preview DataFrame
met_df.head()
```

```python
# Store the Standard Error associated with Met. Sites Grouped by Drug and Timepoint 
met_sites_sem = df.groupby(["Drug", "Timepoint"]).sem()["Metastatic Sites"]

# Convert to DataFrame
met_sites_sem_df = pd.DataFrame(met_sites_sem)
# Preview DataFrame
met_sites_sem_df.head()
```

```python
# Minor Data Munging to Re-Format the Data Frames

met_df = met_df.reset_index()
pivoted_met = met_df.pivot(index='Timepoint', columns='Drug', values='Metastatic Sites')

# Preview that Reformatting worked
pivoted_met.head()
```

```python
# Generate the Plot (with Error Bars)

plt.errorbar(pivoted_met.index, 
             pivoted_met['Capomulin'], 
             yerr=met_sites_sem['Capomulin'].values, 
             linewidth=0.5, 
             marker="o",
             color='r',
             dashes=[2, 1]
            )
plt.errorbar(pivoted_met.index, 
             pivoted_met['Infubinol'], 
             yerr=met_sites_sem['Infubinol'].values, 
             linewidth=0.5, 
             marker="^",
             color='b',
             dashes=[2, 1]
            )
plt.errorbar(pivoted_met.index, 
             pivoted_met['Ketapril'], 
             yerr=met_sites_sem['Ketapril'].values, 
             linewidth=0.5, 
             marker="s",
             color='g',
             dashes=[2, 1]
            )
plt.errorbar(pivoted_met.index, 
             pivoted_met['Placebo'], 
             yerr=met_sites_sem['Placebo'].values, 
             linewidth=0.5, 
             marker="D",
             color='k',
             dashes=[2, 1]
            )

plt.grid()
plt.title('Metastatic Spread During Treatment')
plt.xlabel('Treartment Duration (Days)')
plt.ylabel('Met. Sites')
plt.legend(loc="best", fontsize="small")
# Save the Figure
plt.savefig("MetastaticSpreadDuringTreatment.png", bbox_inches = 'tight')
```

## Survival Rates

```python
# Store the Count of Mice Grouped by Drug and Timepoint (W can pass any metric)
mouse_data = df.groupby(["Drug", "Timepoint"]).count()["Mouse ID"]

# Convert to DataFrame
mouse_df = pd.DataFrame({'Mouse Count' : mouse_data}).reset_index()

# Preview DataFrame
mouse_df.head()
```

```python
# Minor Data Munging to Re-Format the Data Frames

mouse_df = mouse_df.reset_index()
pivoted_mouse_df = mouse_df.pivot(index='Timepoint', columns='Drug', values='Mouse Count')

# Preview that Reformatting worked
pivoted_mouse_df.head()
```

```python
# Generate the Plot (Accounting for percentages)
plt.plot(pivoted_mouse_df.index, 
             pivoted_mouse_df['Capomulin']/pivoted_mouse_df['Capomulin'][0]*100, 
             linewidth=0.5, 
             marker="o",
             color='r',
             dashes=[2, 1]
            )
plt.plot(pivoted_mouse_df.index, 
             pivoted_mouse_df['Infubinol']/pivoted_mouse_df['Infubinol'][0]*100, 
             linewidth=0.5, 
             marker="^",
             color='b',
             dashes=[2, 1]
            )
plt.plot(pivoted_mouse_df.index, 
             pivoted_mouse_df['Ketapril']/pivoted_mouse_df['Ketapril'][0]*100, 
             linewidth=0.5, 
             marker="s",
             color='g',
             dashes=[2, 1]
            )
plt.plot(pivoted_mouse_df.index, 
             pivoted_mouse_df['Placebo']/pivoted_mouse_df['Placebo'][0]*100, 
             linewidth=0.5, 
             marker="D",
             color='k',
             dashes=[2, 1]
            )

plt.grid()
plt.title('Survival During Treatment')
plt.xlabel('Time (Days)')
plt.ylabel('Survival Rate(%)')
plt.legend(loc="best", fontsize="small")
# Save the Figure
plt.savefig("SurvivalDuringTreatment.png", bbox_inches = 'tight')
plt.show()
```
